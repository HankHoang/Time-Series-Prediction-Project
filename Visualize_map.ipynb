```python
#Reading the data
import pandas as pd
df = pd.read_csv('data_0.csv')
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

# Create a Basemap of California
m = Basemap(
    projection='merc',
    llcrnrlat=32.5,
    urcrnrlat=42.5,
    llcrnrlon=-124.75,
    urcrnrlon=-113.75,
    resolution='h'
)
```

# Showing the map of California

```python
# Add coastlines and state borders
m.drawcoastlines()
m.drawstates()

# Plot the data points on the map
m.scatter(df['LON'], df['LAT'], latlon=True, s=10, c='r', alpha=0.5)

# Show the map
plt.show()
```

## Zoom in the main area

```python
# Create a Basemap of Los Angeles
m = Basemap(llcrnrlon=-119.5, llcrnrlat=32.5, urcrnrlon=-114, urcrnrlat=36.5, projection='mill', resolution='l')

# Draw coastlines, countries, and states
m.drawcoastlines()
m.drawcountries()
m.drawstates()

# Fill the land with color
m.fillcontinents(color='#ddaa66', lake_color='#7777ff')

# Plot the data points on the map
m.scatter(df['LON'], df['LAT'], latlon=True, s=10, c='r', alpha=0.5)

# Draw the map boundary
m.drawmapboundary(fill_color='#7777ff')

# Add a title
plt.title('Map of Los Angeles')

# Show the map
plt.show()
```
